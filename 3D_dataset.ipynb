```python
import numpy as np
from PIL import Image
import os

# Function to load an image given the path
def load_image(path):
    try:
        return Image.open(path)
    except Exception as e:
        print(f"Error loading image {path}: {e}")
        return None

# Function to preprocess the image (resize and normalize)
def preprocess_image(img, target_size=(128, 128)):
    # Resize image
    img = img.resize(target_size)
    # Convert image to numpy array and normalize (scale to range [0, 1])
    img_array = np.array(img) / 255.0
    return img_array

# Prepare the data arrays
X = []  # For image data
y = []  # For labels

# Iterate over the DataFrame and load images
for index, row in df.iterrows():
    img = load_image(row['path'])
    if img:
        img_array = preprocess_image(img)
        X.append(img_array)  # Append image to X
        y.append(1 if row['label'] == 'defected' else 0)  # Append label (1 for defected, 0 for non_defected)

# Convert lists to numpy arrays
X = np.array(X)
y = np.array(y)

# Print shapes to verify
print(f"X shape: {X.shape}, y shape: {y.shape}")  # Should print (num_images, 128, 128, 3) for X and (num_images,) for y
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Build the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Sigmoid for binary classification (0 or 1)
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)  # Training with validation split
```

```python
# Evaluate the model (assumes you've split the data into X_test and y_test)
test_loss, test_accuracy = model.evaluate(X, y)  # If you don't have a separate test set, use the full data
print(f"Test Accuracy: {test_accuracy}")
```
